```python
def build_rulebook(rules):
    rulebook = {}
    for rule in rules.split('\n'):
        first, last = list(map(int, rule.split('|')))
        
        if first in rulebook:
            rulebook[first].append(last)
        else:
            rulebook[first] = [last]
    
    return rulebook

def valid(rulebook, update):
    for i, n in enumerate(update):
        rule = rulebook.get(n, [])
        for r in rule:
            if r in update[:i]:
                return False
            
    return True
    

def sort_update(rulebook, update):
    update = update[::-1]
    while not valid(rulebook, update[::-1]):
        
        for i, n in enumerate(update):
            rule = rulebook.get(n, [])
            if not rule:
                continue

            for j in range(i+1, len(update)):
                if update[j] in rule:
                    update[i], update[j] = update[j], update[i]  
    
    return update[::-1]



def a(rules, updates):
    rulebook = build_rulebook(rules)
    res = 0
    for update in updates.split('\n'):
        update = list(map(int, update.split(',')))
        if valid(rulebook, update):
            res += update[len(update) // 2]
    
    return res

def b(rules, updates):
    rulebook = build_rulebook(rules)
    res = 0
    updates = updates.split('\n')
    for i, update in enumerate(updates):
        print(i, '/', len(updates))
        update = list(map(int, update.split(',')))
        if valid(rulebook, update):
            continue

        sorted_update = sort_update(rulebook, update)

        res += sorted_update[len(sorted_update) // 2]
    
    return res


rules, updates = utils.read_file_content_raw('day5').split('\n\n')

b(rules, updates)
```
