```python
grid = utils.read_file_content_raw('day6')
grid = list(map(list, grid.split()))

pos = None

for i in range(len(grid)):
    for j in range(len(grid[i])):
        if grid[i][j] == "^":
            pos = (i,j)
```

```python
def a(grid, pos):
    dir = ['u', 'r', 'd', 'l']
    moves = {'u' : (-1,0), 'r' : (0,1), 'd' : (1,0), 'l' : (0,-1)}
    cdir = dir[0]
    visited = set([pos])
    while True:
        nextpos = (pos[0] + moves[cdir][0], pos[1] + moves[cdir][1])

        if not (0 <= nextpos[0] < len(grid) and 0 <= nextpos[1] < len(grid[0])):
            break

        
        next = grid[nextpos[0]][nextpos[1]]

        if next == "#":
            cdir = dir[(dir.index(cdir) + 1) % 4]
        else:
            pos = nextpos
            grid[pos[0]][pos[1]] = "|"


        visited.add(pos)
            
    
    return len(visited), grid


def b(grid, pos):
    dir = ['u', 'r', 'd', 'l']
    moves = {'u' : (-1,0), 'r' : (0,1), 'd' : (1,0), 'l' : (0,-1)}
    cdir = dir[0]
    visited = set()
    while True:
        nextpos = (pos[0] + moves[cdir][0], pos[1] + moves[cdir][1])

        if (nextpos, cdir) in visited or not (0 <= nextpos[0] < len(grid) and 0 <= nextpos[1] < len(grid[0])):
            print((nextpos, cdir) in visited)
            break

        
        next = grid[nextpos[0]][nextpos[1]]

        if next == "#":
            cdir = dir[(dir.index(cdir) + 1) % 4]
        else:
            pos = nextpos
            grid[pos[0]][pos[1]] = "|"


        visited.add((pos, cdir))
            
    
    return (nextpos, cdir) in visited


stuck = []
for i in range(len(grid)):
    for j in range(len(grid[i])):
        print()
        if (i,j) == pos or (grid[i][j] == "#"):
            continue

        gc = np.array(grid)

        gc[i][j] = "#"

        stuck.append(b(gc, pos))

print(sum(stuck))
```
